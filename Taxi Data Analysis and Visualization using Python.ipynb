```python
#Taxi Data Analysis and Visualization using Python

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the 'taxis' dataset
df = sns.load_dataset("taxis")
df
```

```python
df.info()
```

```python
#Handling Missing Values
#Check for missing values in the dataset and identify columns with missing data.
df
```

```python
df.isnull().sum()
```

```python
#Identifying missing values
import warnings
warnings.filterwarnings('ignore')
df['payment'].fillna(df['payment'].mode()[0], inplace=True)
df['pickup_zone'].fillna(df['pickup_zone'].mode()[0], inplace=True)
df['dropoff_zone'].fillna(df['dropoff_zone'].mode()[0], inplace=True)
df['pickup_borough'].fillna(df['pickup_borough'].mode()[0], inplace=True)
df['dropoff_borough'].fillna(df['dropoff_borough'].mode()[0], inplace=True)
df.isnull().sum()
```

```python
#Visualizations using Matplotlib/Pandas Plot
#Line Chart
#Fare price over time

plt.figure(figsize=(20,10))
df = df.sort_values(by='pickup')
plt.plot(df['pickup'], df['fare'])

plt.xlabel("Pickup Time")
plt.ylabel("Fare Amount")
plt.title("Fare Over Time")

plt.show()
```

```python
#Bar Chart :Total fare for each pickup_borough
plt.figure(figsize=(10,5))
df.groupby('pickup_borough')['fare'].sum().plot(kind='bar')
plt.title("Total Fare for each Pickup Borough")

plt.xlabel("Pickup Borough")
plt.ylabel("Total Fare")
plt.show()
```

```python
#pie chart
plt.figure(figsize=(10,5))
plt.title("Trip Payment Mode")
df['payment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show
```

```python
plt.figure(figsize=(10,6))

plt.hist(df['distance'], bins=30,color='orange')

plt.xlabel("Distance")
plt.ylabel("Frequency")
plt.title("Distribution of Trip Distance")

plt.show()
```

```python
#box-plot
df.boxplot(column='tip',by='pickup_borough',color='red')
plt.title("Tip Distribution by Pickup Borough")
plt.suptitle("")
plt.xlabel("Pickup Borough")
plt.ylabel("Tip Amount")

plt.show()
```

```python
#Visualizations using Seaborn
#Count Plot
import seaborn as sns
plt.figure(figsize=(10,5))
sns.countplot(x='pickup_borough',data=df,color='green')
plt.title("Count of trips by pickup borough")
plt.xlabel("Pickup Borough")
plt.ylabel("Count")
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x='distance',y='fare',data=df,color='red')

plt.title("Distance vs Fare")
plt.xlabel("Distance")
plt.ylabel("Fare")
plt.show()
```

```python
data = df[['distance', 'fare', 'tip', 'tolls', 'total']]
corr_matrix=data.corr()
sns.heatmap(data.corr(),annot=True,cmap="coolwarm")
plt.title("Taxi Trip Feature Correlation")
plt.show()
```

```python
pair_pl=df[['distance','fare','tip','total','pickup_zone']]
sns.pairplot(pair_pl)
plt.suptitle("Pairwise Relationship Between Distance, Fare, Tip and Total",y=1.02)
plt.show
```

```python
#ViolinePlot
plt.figure(figsize=(10,5))
sns.violinplot(x='payment',y='fare',data=df,color='purple')

plt.title("fare distribution by payment method")
plt.xlabel("Payment method")
plt.ylabel("Fare")
plt.show()
```
